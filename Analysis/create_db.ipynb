```python

import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
database_path = "../data/emission_population.sqlite"
```

```python
engine=create_engine(f"sqlite:///{database_path}", echo=True)
conn=engine.connect()
```

```python
conn.execute("DROP TABLE emission_population")
```

```python
# Create the table - note that if we use automap_base, the table must have primary key!
conn.execute("""
    CREATE TABLE IF NOT EXISTS emission_population (
        City_State text,
        City text PRIMARY KEY,
        State text ,
        population2010 integer,
        population2011 integer,
        population2012 integer,
        population2013 integer,
        population2014 integer,
        population2015 integer,
        population2016 integer,
        population2017 integer,
        emissions2010 real,
        emissions2011 real,
        emissions2012 real,
        emissions2013 real,
        emissions2014 real,
        emissions2015 real,
        emissions2016 real,
        emissions2017 real,
        lat decimal(9,6),
        lng decimal(8,6)
        
    )
""")
```

```python
df= pd.read_csv("../data/emission_population_coordinates.csv")
df.head()
```

```python
df.columns
```

```python
df = df.drop(["Unnamed: 0"], axis=1)
df
```

```python
# insert data into the DB
df.to_sql("emission_population", conn, if_exists="append", index=False)
```
