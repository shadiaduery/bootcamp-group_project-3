```python
import pandas as pd
```

```python
# The path to our CSV file
file = "Resources/data/Population2010_2019.csv"
```

```python
# Read population data into pandas
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file, encoding="ISO-8859-1")
df.head(10)
```

```python
# Get a list of all of our columns for easy reference
df.columns
```

```python
# Drop all rows with missing information
df = df.dropna(how='any')
```

```python
 # Extract what we need for appropriate columns
reduced_population_df = df.loc[:, ["Unnamed: 1", "Unnamed: 2", "Unnamed: 5",
                                    "Unnamed: 6", "Unnamed: 7", "Unnamed: 8", 
                                    "Unnamed: 9", "Unnamed: 10","Unnamed: 11", 
                                    "Unnamed: 12", "Unnamed: 13"]]
reduced_population_df
```

```python
reduced_population_df = reduced_population_df.rename(columns={"Unnamed: 1": "City, State",
                                        "Unnamed: 2": "2010", "Unnamed: 6": "2012",
                                        "Unnamed: 5": "2011", "Unnamed: 7": "2013",
                                        "Unnamed: 8": "2014", "Unnamed: 9": "2015",
                                        "Unnamed: 10": "2016", "Unnamed: 11": "2017",
                                        "Unnamed: 12": "2018", "Unnamed: 13": "2019",
                                        "Reporter\nReal\nName": "Reporter Real Name",})
                                        
reduced_population_df
```

```python
reduced_population_df.count()
```

```python
# print cleaned up data for merging to csv
reduced_population_df.to_csv("population_cleaned.csv", encoding="utf-8", index=False)
```
