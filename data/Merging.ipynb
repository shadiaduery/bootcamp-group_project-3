```python
import pandas as pd
```

```python
df_population = pd.read_csv("population_state.csv")
df_population
```

```python
df_emissions = pd.read_csv("Data_urban_emissions.csv")
df_emissions
```

```python
df_population = df_population.rename(columns={"City, State": "City_State",
                             "2010": "population2010",
                              "2011": "population2011",
                              "2012": "population2012",
                              "2013": "population2013",
                              "2014": "population2014",
                              "2015": "population2015",
                              "2016": "population2016",
                              "2017": "population2017"})
df_population
```

```python
df_population = df_population.drop(["2018", "2019"], axis=1)
df_population
```

```python
df_population.dtypes
```

```python
df_population["population2010"] = df_population["population2010"].str.replace(',', '').astype(int)
df_population["population2011"] = df_population["population2011"].str.replace(',', '').astype(int)
df_population["population2012"] = df_population["population2012"].str.replace(',', '').astype(int)
df_population["population2013"] = df_population["population2013"].str.replace(',', '').astype(int)
df_population["population2014"] = df_population["population2014"].str.replace(',', '').astype(int)
df_population["population2015"] = df_population["population2015"].str.replace(',', '').astype(int)
df_population["population2016"] = df_population["population2016"].str.replace(',', '').astype(int)
df_population["population2017"] = df_population["population2017"].str.replace(',', '').astype(int)

df_population
```

```python
df_emissions = df_emissions.rename(columns={"Sum of kgco2_2010": "emissions2010",
                              "Sum of kgco2_2011": "emissions2011",
                              "Sum of kgco2_2012": "emissions2012",
                              "Sum of kgco2_2013": "emissions2013",
                              "Sum of kgco2_2014": "emissions2014",
                              "Sum of kgco2_2015": "emissions2015",
                              "Sum of kgco2_2016": "emissions2016",
                              "Sum of kgco2_2017": "emissions2017"})
df_emissions
```

```python
emission_population_data = df_population.merge(df_emissions, how="inner", on=["City"])
emission_population_data
```

```python
emission_population_data = emission_population_data.rename(columns={"State_x": "State"})
emission_population_data
```

```python
emission_population_data = emission_population_data.drop(["State_Name", "State_y"], axis=1)
emission_population_data 
```

```python
emission_population_data = emission_population_data [["City_State","City","State",
                                                      "population2010", "population2011", "population2012",
                                                     "population2013", "population2014", "population2015",
                                                      "population2016", "population2017", "emissions2010",
                                                      "emissions2011", "emissions2012", "emissions2013",
                                                      "emissions2014", "emissions2015", "emissions2016",
                                                      "emissions2017"]]
emission_population_data 
```

```python
emission_population_data.to_csv("emission_population_data.csv", encoding="utf-8", index=False)
```
